## Gather Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import json
import wptools
import tweepy
import requests
```

```python
#csv file
twitter = pd.read_csv('twitter-archive-enhanced.csv')
#tsv file
url =  "https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"
r = requests.get(url)
with open('image-predictions.tsv', 'wb') as f:
    f.write(r.content)
image = pd.read_csv('image-predictions.tsv', sep='\t')
```

```python
#twitter api
#https://stackoverflow.com/questions/18314749/setup-for-python-tweepy
consumer_key = "x"
consumer_secret = "x"
access_token = "x"
access_token_secret = "x"
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
#https://stackoverflow.com/questions/24002536/get-tweepy-search-results-as-json/29006599
#https://stackoverflow.com/questions/4775749/is-there-a-way-to-get-a-tweets-favorite-count
tweet_id = image['tweet_id']
tweets = [] # tweets that can be found by id
dne = [] #tweets that can't be found by id
new_api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True, parser = tweepy.parsers.JSONParser())

for i in tweet_id:
    try:
        status = new_api.get_status(i)
        faves = status['favorite_count'] #favorites on the tweet
        retweets = status['retweet_count'] #retweets on the tweet
        #datetime = pd.to_dataframe(pg['created_at'])
        tweets.append({'tweet_id':i, 'faves':faves, 'retweets':retweets}) #new dataframe
    except Exception as e:
        dne.append(i) 
```

```python
#creating new dataframe
json = pd.DataFrame(tweets, columns = ['tweet_id', 'faves', 'retweets'])
#creating csv file for dataframe
json.to_csv('tweet_json.txt')
new_df = pd.read_csv('tweet_json.txt')
#an extra index was created that is not needed
new_df.drop(columns = ['Unnamed: 0'], inplace=True)
new_df.head()
```

## Tidiness
**Define: 1 . Merging all three dataframes together**

```python
df_clean = pd.merge(left = df_clean, right = image_clean, how = 'inner', right_on = 'tweet_id', left_on = 'tweet_id')
```

**Test:**

**Define: 2. Create only one rating column by taking the rating_numerator/rating_denominator**

**Code:**

```python
rating = df_clean['rating_numerator']/df_clean['rating_denominator']
rating
```

```python
del df_clean['rating_numerator']
del df_clean['rating_denominator']
df_clean['rating'] = rating
df_clean['rating']
```

```python
#some of these rating do not proportionally make sense, so deleting those rows
df_clean['rating'].nlargest(7)
```

```python
#dropping rows with ratings greater than 20/10
df_clean.drop(df_clean[df_clean['rating'] > 2].index, inplace = True)
df_clean['rating'].nlargest(5)
```

**Test:**

```python
df_clean.head()
```

**Define: 3. Combine the doggo, floofer, pupper, and puppo into one column**

**Code:**

```python
#https://stackoverflow.com/questions/41449555/pandas-combine-two-columns-with-null-values
dogtionary = df_clean['doggo']
df_clean['dogtionary'] = dogtionary
multiple = df_clean['doggo']+df_clean['pupper']+df_clean['puppo']+df_clean['floofer']
df_clean['multiple'] = multiple
```

```python
df_clean['dogtionary'].fillna(df_clean['floofer'], inplace=True)
df_clean['dogtionary'].fillna(df_clean['pupper'], inplace=True)
df_clean['dogtionary'].fillna(df_clean['puppo'], inplace=True)
```

```python
del df_clean['doggo']
del df_clean['floofer']
del df_clean['pupper']
del df_clean['puppo']
```
